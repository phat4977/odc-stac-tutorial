### Display default map

Default map : lat = 13.726 and lon = 100.514  (Bangkok latitude/longitude)

```python
from dream_river.plotimg import show_map

Map = show_map()
Map
```

### Assign zoom level

```python
Map = show_map(zoom=15)
Map
```

### Display map from lat/lon

```python
lat = (14.726)  # Philipine lat/lon
lon = (121.0)

show_map(lat,lon, zoom = 15)
```

### Display map from bounding box

```python
# Query bounding box
lat_range = (9.135, 11.106)
lon_range = (98.812, 100.821)
```

```python
from dream_river.plotimg import display_map

display_map(lon_range,lat_range)
```

### Display a vector data

```python
from dream_river.plotimg import show_vector

path=['training/urban_sawan.geojson']

show_vector(path)
```

### Display multi vector data

```python
files=['training/green_area_sawan.geojson','training/urban_sawan.geojson','training/water_sawan.geojson']

show_vector(files)
```

### Display by Geopandas

```python
import geopandas as gpd

gdf = gpd.read_file('training/urban_sawan.geojson')

gdf.head()
```

```python
gdf.explore(column='class', legend=False)
```
