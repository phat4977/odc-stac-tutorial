## Import Package

```python
import geopandas as gpd
from dream_river.plotimg import show_vector
from dream_river.geobox import convert_geojson_CRS, convert_shp_CRS
```

## Explore data

The interactive map of Dream-river package is use EPSG:4326 but the available dataset on this datacube is EPSG:32647.

Thus, we need to convert the CRS from EPSG:4326 to EPSG:32647.

```python
path=['training/urban_area.geojson']

show_vector(path)
```

```python
gdf=gpd.read_file('training/urban_area.geojson')
gdf.crs
```

AS you see above, The Geographic Coordinate System of this file is EPSG:4326, but the available dataset on this datacube is EPSG:32647. 

Thus, we need to convert the CRS from EPSG:4326 to EPSG:32647.

## Convert CRS

This can convert CRS in both Geojson and Shapefile file format by using:
- convert_geojson_CRS
- covert_shp_CRS

```python
# set input output path
input_path = 'training/urban_area.geojson'
output_path = 'training/urban_area32647.geojson'

# function to convert CRS (EPSG: 32647)
convert_geojson_CRS(input_path, output_path, crs=32647)
```

## Result

As you can see, the result show the different CRS from The input file.

```python
result=gpd.read_file(output_path)
result.crs
```

```python
result.explore()
```

As the previouse explanation, The interactive map of Dream-River library use CRS= EPSG:4326.

Thus, after convert CRS to EPSG:32647, you can use interactive map of Geopandas library to explore output as the code above.
